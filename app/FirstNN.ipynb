```python
# Build 3 neuron having 3 by 3 matrix of weights

inputs = [1.3, 2.4, 3.6]

weights1 = [0.2, 0.8, -0.5]
weights2 = [0.5, -0.84, 0.53]
weights3 = [-0.21, -0.62, 0.16]

bias1 = 2
bias2 = 4
bias3 = 0.8

# Neuron 1
output = [ inputs[0] * weights1[0] + 
          inputs[1] * weights1[1] +
          inputs[2] * weights1[2] + bias1,
# Neuron 2
    inputs[0] * weights2[0] + 
    inputs[1] * weights2[1] +
    inputs[2] * weights2[2] + bias2,
# Neuron 3
    inputs[0] * weights3[0] + 
    inputs[1] * weights3[1] +
    inputs[2] * weights3[2] + bias3
]

print(output)
```

```python
inputs = [1.3, 2.4, 3.6, -1.8]
weights = [[0.2, 0.8, -0.5, 0.3],[0.5, -0.84, 0.53, -0.3],[-0.21, -0.62, 0.16, 0.78]]
bias = [2, 4, 0.8]
layer_outputs = []
for neuron_weights, neuron_bias in zip(weights, bias):
    neuron_output = 0
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input * weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)
print(layer_outputs)

```

```python
# creating 4 neuron
inputs = [1.3, 2.4, 3.6, -1.8]

weights = [[0.2, 0.8, -0.5, 0.3],
           [0.5, -0.84, 0.53, -0.3],
           [-0.21, -0.62, 0.16, 0.78],
           [0.14, -0.45, 0.62, -0.19]]

bias = [2, 4, 0.8, -1.2]

# for loop to dot product weights and inputs

layer_outputs = []

for neuron_weights, neuron_bias in zip(weights, bias):
    neuron_output = 0
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input * weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)
    
print(layer_outputs)

```

```python
# creatig a single neuron
input = [1, 2, 3, 4]
weight = [0.2, 0.8, -0.5, 0.5]
bias = 2

# dot product the input and weight matrix with numpy
output = np.dot(input, weight) + bias
print(output)
```

```python
# creating a layer having 4 inputs
inputs = [
    [1.3, 2.4, 3.6, -1.8],
    [2.3, 4.5, -6.7, 1.2],
    [-2.3, 4.5, 6.7, -1.2],
    [1.2, 2.3, 4.5, 6.7]]
weights = [[0.2, 0.8, -0.5, 0.3],
           [0.5, -0.84, 0.53, -0.3],
           [-0.21, -0.62, 0.16, 0.78],
           [0.14, -0.45, 0.62, -0.19]] 
bias = [2, 4, 0.8, -1.2]

output = np.dot(inputs, np.array(weights).T) + bias 

print(output)
```

```python
# creating 2 layers having 4 inputs

inputs = [1.3, 2.4, 3.6, -1.8]

weights1 = [[0.2, 0.8, -0.5, 0.3],
           [0.5, -0.84, 0.53, -0.3],
           [-0.21, -0.62, 0.16, 0.78],
           [0.14, -0.45, 0.62, -0.19]]
bias1 = [2, 4, 0.8, -1.2]


weights2 = [[0.2, 0.8, -0.5, 0.3],
           [0.5, -0.84, 0.53, -0.3],
           [-0.21, -0.62, 0.16, 0.78],
           [0.14, -0.45, 0.62, -0.19]]
bias2 = [2, 4, 0.8, -1.2]

output1 = np.dot(inputs, np.array(weights1).T) + bias1
# apply Relu activation function
output1 = np.maximum(0, output1)
print("Output for the first layer: ",*output1, sep=', ')

output2 = np.dot(output1, np.array(weights2).T) + bias2
# apply Relu activation function
output2 = np.maximum(0, output2)
print("Output for the second layer: ",*output2, sep=', ')




```
